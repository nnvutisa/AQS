```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

Importing Bokeh

```python
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.layouts import widgetbox
from bokeh.models.widgets import CheckboxButtonGroup, Panel, Tabs
from bokeh.models import ColumnDataSource, DataRange1d, Select, CustomJS, HoverTool
output_notebook()
```

Read Data.

```python
city_aqi = pd.read_csv('city_aqi.csv',index_col=0)
```

```python
city_aqi.head()
```

```python
city_grouped = city_aqi.groupby('city_name')
```

## Plot

```python
from bokeh.palettes import Set3
cm = Set3[10]
```

```python
p2 = figure(
    width = 900,
    tools="hover",
    title="City Air Quality Index",
    x_axis_label='Year', y_axis_label='AQI',
    #y_range=[0,100]
)

i=0
for name,group in city_grouped:
    #data
    x = group.sort_values(by='year')['year']
    y = group.sort_values(by='year')['Ozone']
    
    #plot
    s1 = p2.line(x,y,legend=name,line_color=cm[i],line_width=2)
    s2 = p2.circle(x,y,size=8,legend=name,line_color=cm[i],fill_color=cm[i])
    p2.add_tools(HoverTool(renderers=[s2], tooltips= {"City":name, "AQI":"$y{0.0}"} ))
    
    i +=1
    if i >= 9:
        break
        
p2.legend.click_policy="hide"
```

```python
p1 = figure(
    width = 900,
    tools="hover",
    title="City Air Quality Index",
    x_axis_label='Year', y_axis_label='AQI',
    #y_range=[0,100]
)

i=0
for name,group in city_grouped:
    #data
    x = group.sort_values(by='year')['year']
    y = group.sort_values(by='year')['Carbon monoxide']
    
    #plot
    s1 = p1.line(x,y,legend=name,line_color=cm[i],line_width=2)
    s2 = p1.circle(x,y,size=8,legend=name,line_color=cm[i],fill_color=cm[i])
    p1.add_tools(HoverTool(renderers=[s2], tooltips= {"City":name, "AQI":"$y{0.0}"} ))
    
    i +=1
    if i >= 9:
        break
        
p1.legend.click_policy="show"
```

```python
tab1 = Panel(child=p1, title="Carbon Monoxide")
tab2 = Panel(child=p2, title="Ozone")

tabs = Tabs(tabs=[ tab1, tab2 ])

show(tabs)
```
